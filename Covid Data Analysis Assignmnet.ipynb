## Q3. Maximum number of people vaccinated in a country

```python
def max_vacc (country):
    return int(df[df['country'] == country].max()['people_vaccinated'])
```

```python
data = []
country = df['country'].unique()
for i in country:
    max_vacc(i)
    data.append([i,max_vacc(i)])
    
```

```python
data
```

## Q4. Top 5 Countries in terms of number of people vaccinated

```python
data_frame = pd.DataFrame(data, columns =['country', 'people_vaccinated'])
```
